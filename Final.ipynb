```python
def barplot(df, name):
    name = str(name)
    fig = plt.figure(figsize=(18,6))
    sns.barplot(x=df.word[:10], y=df.number[:10], palette=sns.cubehelix_palette(reverse = True))
    plt.title(f'Top 10 Most Common Words used in {name} Articles')
    plt.show()
```
